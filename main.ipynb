```python
import osmnx as ox
import networkx as nx
```

```python
G = ox.graph_from_place('Ciudad de México, México', network_type='drive')
 
```

```python
origen = ox.geocode('19.4326, -99.1332') 
destino = ox.geocode('19.4345, -99.1507')  
origen_node = ox.distance.nearest_nodes(G, X=origen[1], Y=origen[0])
destino_node = ox.distance.nearest_nodes(G, X=destino[1], Y=destino[0])

ruta = nx.shortest_path(G, origen_node, destino_node, weight='length')
fig, ax = ox.plot_graph_route(G, ruta, route_linewidth=6, node_size=0, bgcolor='k')
```

```python
longitud_total_metros = sum([G[u][v][0]['length'] for u, v in zip(ruta[:-1], ruta[1:])])
longitud_total_metros
```
